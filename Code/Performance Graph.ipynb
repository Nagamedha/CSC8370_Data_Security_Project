```python
import matplotlib.pyplot as plt

# Accuracy data for Level 1 and Level 2
level1_epochs = list(range(1, 21))
level1_accuracies = [
    98.14, 98.65, 98.82, 99.04, 98.95, 99.11, 99.17, 99.14, 99.28, 99.25,
    99.26, 99.29, 99.28, 99.23, 99.30, 99.28, 99.35, 99.32, 99.28, 99.36
]

level2_rounds = list(range(1, 11))
level2_accuracies = [
    94.51, 96.65, 96.65, 97.23, 97.61, 97.78, 97.89, 97.96, 97.94, 98.01
]

# Create the combined graph
plt.figure(figsize=(10, 6))
plt.plot(level1_epochs, level1_accuracies, label='Level 1: Centralized Learning', marker='o')
plt.plot(level2_rounds, level2_accuracies, label='Level 2: Federated Learning', marker='s')

# Add labels, title, and legend
plt.xlabel('Epochs (Level 1) / Rounds (Level 2)', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.title('Accuracy Comparison: Centralized vs Federated Learning', fontsize=14)
plt.legend(fontsize=12)

# Annotate the data points
for i, acc in enumerate(level1_accuracies):
    plt.text(level1_epochs[i], acc, f"{acc}%", fontsize=8, ha='right')
for i, acc in enumerate(level2_accuracies):
    plt.text(level2_rounds[i], acc, f"{acc}%", fontsize=8, ha='left')

# Show grid and plot
plt.grid(alpha=0.3)
plt.show()
```
